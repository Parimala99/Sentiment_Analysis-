# POS Tagger for Malayalam Text

```python
import nltk
paragraph="നിങ്ങളെ എല്ലാവരേയും പോലെ ആക്കാൻ ശ്രമിക്കുന്ന ഒരു ലോകത്ത് നിങ്ങളായിരിക്കുക എന്നതാണ് ജീവിതത്തിലെ ഏറ്റവും വലിയ വെല്ലുവിളി. ഗവേഷണ ആവശ്യത്തിനായി ശാസ്ത്രജ്ഞർ 5 കുരങ്ങുകളെ ഒരു കൂട്ടിൽ വച്ചു, നടുക്ക് ഒരു ഗോവണി, മുകളിൽ വാഴപ്പഴം. ഒരു കുരങ്ങ് ഗോവണിയിലേക്ക് പോകുമ്പോഴെല്ലാം ശാസ്ത്രജ്ഞൻ ബാക്കി കുരങ്ങുകളെ തണുത്ത വെള്ളത്തിൽ കുതിർത്തു. കുറച്ചു കഴിഞ്ഞപ്പോൾ, ഒരു കുരങ്ങൻ ഗോവണിയിൽ കയറാൻ ശ്രമിക്കുമ്പോഴെല്ലാം മറ്റുള്ളവർ അവനെ അടിക്കും. കുറച്ച് സമയത്തിനുശേഷം, ഒരു കുരങ്ങനും പ്രലോഭനം കണക്കിലെടുക്കാതെ ഗോവണിയിലേക്ക് കയറാൻ തുനിഞ്ഞില്ല. കുരങ്ങുകളിലൊന്ന് മാറ്റിസ്ഥാപിക്കാൻ ശാസ്ത്രജ്ഞർ തീരുമാനിച്ചു."
sentence=nltk.sent_tokenize(paragraph)
print(sentence)
```

```python
from nltk import word_tokenize, pos_tag
pos_tag(word_tokenize(paragraph))
```
